### Usando "|" para separ valores de questões de múltipla escolha

```python
def treatMulti(value, question):
    choices = value.split(",")
    choices = " ".join([v.lstrip().rstrip().replace(" ", "_")+"_"+question for v in choices])
    return choices
```

```python
# Q4, Q9, Q25
df["Q4_V2"] = [treatMulti(choices, "Q4") for choices in df.Q4]    
df["Q9_V2"] = [treatMulti(choices, "Q9") for choices in df.Q9]  
df["Q25_V2"] = [treatMulti(choices, "Q25") for choices in df.Q25]  
```
